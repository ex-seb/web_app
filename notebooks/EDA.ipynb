```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("metro.csv")
df.info()
df.sample(10)
```

```python
linea_a = df[df['linea'].str.contains('Linea A')]
linea_a.head()
```

```python
fig = px.histogram(linea_a, x="estacion", y='afluencia')
fig.show()
```

```python
fig = px.scatter(df, x="estacion", y="tipo_pago")
fig.show()
```

```python
fig = px.bar(linea_a, x="afluencia", y="estacion",color="tipo_pago", orientation='h')
fig.show()
```
